## Data Handling

```python
# find missing values
df.isna().sum()
```

```python
# replace null names columns
df['name']=df['name'].fillna('unknown')
```

```python
df.isna().sum()
```

```python
# replace ratings 
df['rating']=df['rating'].ffill()
# reviews                 
df['reviews']=df['reviews'].ffill()
# description                
df['description']=df['description'].ffill()
# ceo_approval            
df['ceo_approval']=df['ceo_approval'].ffill()
# ceo_count               
df['ceo_count']=df['ceo_count'].ffill()
# interview_experience    
df['interview_experience']=df['interview_experience'].ffill()
# interview_difficulty    
df['interview_difficulty']=df['interview_difficulty'].ffill()
# interview_duration     
df['interview_duration']=df['interview_duration'].ffill()
# interview_count         
df['interview_count']=df['interview_count'].ffill()
# headquarters            
df['headquarters']=df['headquarters'].ffill()
# employees               
df['employees']=df['employees'].ffill()
# industry                
df['industry']=df['industry'].ffill()
# revenue                 
df['revenue']=df['revenue'].ffill()
# website                  
df['website']=df['website'].ffill()
```

```python
# replace ratings 
df['rating']=df['rating'].bfill()
# reviews                 
df['reviews']=df['reviews'].bfill()
```
