Seaborn is primarily a data visualization library built on top of Matplotlib and tightly integrated with pandas. 

```python
import seaborn as sns
df=sns.load_dataset('tips')
df
```

```python
# to get built in dataset in seaborn
sns.get_dataset_names()
```

```python
#to find correlation 
df.select_dtypes(['number']).corr()
```

```python
# convert to heatmap
sns.heatmap(df.select_dtypes(['number']).corr())
```

```python
# to plot two diff graph in a single graph
sns.jointplot(data=df, x='total_bill', y='tip')
```

```python
# graphs comparing each numerical columns
sns.pairplot(df)
```

```python
# countplot is similar to value count
sns.countplot(x='sex',data=df)
# set style
sns.set_style('darkgrid')
```

```python
# 
sns.countplot(x='time',data=df,hue='smoker')
```

```python
sns.barplot(x='total_bill',y='smoker',data=df,hue='sex')
```

```python
# to detect outliers(data outside our range)
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
import seaborn as sns
df=sns.load_dataset('titanic')
df
```

```python
df.select_dtypes(['number']).corr()
```

```python
sns.heatmap(df.select_dtypes(['number']).corr())
```

```python
sns.jointplot(data=df, x='class', y='survived')
```

```python
sns.pairplot(df)
```

```python
# countplot is similar to value count
sns.countplot(x='class',data=df)
# set style
sns.set_style('darkgrid')
```

```python
# class v/s survived
sns.countplot(x='class',data=df,hue='survived')
```

```python
sns.barplot(x='survived',y='class',data=df,hue='sex')
```

```python
# to detect outliers(data outside our range)
sns.boxplot(x='survived',y='fare',data=df)
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
# X categorical value
# y continous value
sns.violinplot(x='species',y='sepal_length',data=df)
```

```python
# jitter,dodge - not to overlap the points
# hue must be categorical value
sns.stripplot(x='species',y='sepal_length',data=df,jitter=True,hue='petal_width',dodge=True)
```

```python
# KDE is the smooth curve - kernal density estimator
# histplot - only plot one single column
import numpy as np
sns.set(style='white')
rs=np.random.RandomState(10)
d=rs.normal(size=100)
sns.histplot(d,kde=True,color='g')
```

```python
# lineplot
sns.lineplot(x='species',y='petal_length',hue='petal_width',style='sepal_length',data=df)
```

```python
sns.lmplot(x='sepal_length',y='petal_length',data=df)
```
