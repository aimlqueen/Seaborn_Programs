```python
data['test_score(out of 10)'].fillna(data['test_score(out of 10)'].mean(),inplace=True)
```

```python
experience_data={
    'one':1,
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'seven':7,
    'eight':8,
    'nine':9,
    'ten':10,
    'eleven':11
}    
```

```python
data['experience']=data['experience'].replace(experience_data).astype(float)
```

```python
data['experience']=data['experience'].fillna(data['experience'].mean())
```
